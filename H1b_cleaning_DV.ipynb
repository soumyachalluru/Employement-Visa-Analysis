```python
import pandas as pd

# Try a different encoding
try:
    data = pd.read_csv("h1b_disclosure_data_2015_2019.csv", encoding='latin-1')
except Exception as e:
    print(f"An error occurred: {e}")
else:
    df = pd.DataFrame(data)

```

```python
df.shape
```

```python
df
```

```python
# Identify the row with the specified CASE_NUMBER
row_index = df[df['CASE_NUMBER'] == 'I-200-18282-647333'].index[0]

# Swap SOC_NAME and SOC_CODE for the specified row
df.loc[row_index, ['SOC_NAME', 'SOC_CODE']] = df.loc[row_index, ['SOC_CODE', 'SOC_NAME']].values
```

```python
df
```

```python
null_count_per_column = df.isnull().sum()  # Count of null values in each column
print("\nNumber of null values per column:")
print(null_count_per_column)
```

```python
df = df.dropna(subset=['EMPLOYER_NAME', 'WORKSITE_CITY','SOC_NAME','JOB_TITLE','FULL_TIME_POSITION','PREVAILING_WAGE'])
```

```python
null_count_per_column = df.isnull().sum()  # Count of null values in each column
print("\nNumber of null values per column:")
print(null_count_per_column)
```

```python
df = df.drop(columns="WORKSITE")
```

```python
df.columns
```

```python
df.shape
```

```python
# Remove comma and everything after it in each SOC_NAME value
df['SOC_NAME'] = df['SOC_NAME'].str.split(',').str[0]
```

```python
print(df['SOC_NAME'].value_counts())
```

```python
# Display basic information about the dataset
print("Basic information about the dataset:")
print(df.info())
```

```python
# Display summary statistics for numerical columns
print("\nSummary statistics for numerical columns:")
print(df.describe())
```

```python

# Perform value counts for categorical columns
print("\nValue counts for categorical columns:")
for column in df.select_dtypes(include=['object']):
    print("\nColumn:", column)
    print(df[column].value_counts())
```
